Some API requests for various transit/traffic apis we'll be using throughout the project

```python
import requests
import json
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# MBTA Monthly Ridership
url = 'https://services1.arcgis.com/ceiitspzDAHrdGO1/arcgis/rest/services/MBTA_Monthly_Ridership_By_Mode/FeatureServer/0/query?outFields=*&where=1%3D1&f=geojson'
ridership_response = requests.get(url).text
ridership_json = json.loads(ridership_response)
ridership_df = pd.json_normalize(ridership_json['features'])
ridership_df
```

```python
# MassDOT Traffic Data
url = 'https://gis.massdot.state.ma.us/arcgis/rest/services/Roads/VMT/FeatureServer/9/query?outFields=*&where=1%3D1&f=geojson'
traffic_response = requests.get(url).text
traffic_json = json.loads(traffic_response)
traffic_df = pd.json_normalize(traffic_json['features'])
traffic_df[['properties.Urban_Type', 'properties.City', 'properties.County', 'properties.AADT', 'properties.From_Date']]
```
